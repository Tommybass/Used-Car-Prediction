### Car Prediction using Engine size feature (Linear Regression)

```python
# extract the input and output feature
x = car_df['Engine size']
y= car_df['Price']
```

```python
# Import relevant libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, r2_score, mean_squared_error
```

```python
# Convert X to 2 dimensional array
x = x.to_numpy().reshape(-1,1)

# Split the data into train and test data(80% , 20% respectively)
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2, random_state= 45)
```

```python
# check the number of data points in the training and testing datasets
print(f'There are {len(x_train)} data points in the training set')
print(f'There are {len(x_test)} data points in the testing set')
```

##### Data Preprocessing

```python
# Scale the training dataset
scale =StandardScaler()

# fit it to x_train
scale.fit(x_train)

# transform the x_train and x_test set
x_train_scaled = scale.transform(x_train)
x_test_scaled = scale.transform(x_test)
```

#### Linear Regression model training

```python
# store linear regression in a variable
car_linear = LinearRegression()

# fit the model to the training data
car_linear.fit(x_train_scaled, y_train)
```

```python
# Get the fit coefficient and gradient from the model
gradient = car_linear.coef_[0]
intercept = car_linear.intercept_

# Print the coefficient
print(f'Gradient: {gradient}')
print(f'Intercept: {intercept}')
```

#####  Evaluate the model

```python
# predict the output of the testing set
car_pred = car_linear.predict(x_test_scaled)

# calculate mean absolute error
mean_absolute = mean_absolute_error(y_test, car_pred)

# calculate the R2 score
R2 = r2_score(y_test, car_pred)

# print both metrics
print(f'Mean Absolute Error: {mean_absolute}')
print(f'R2: {R2}')
```

```python
x_fit = np.arange(1,5,1).reshape(-1,1)
x_fit_scaled = scale.transform(x_fit)
y_fit = car_linear.predict(x_fit_scaled)
sns.scatterplot(car_df, x='Engine size', y='Price')
plt.plot(x_fit,y_fit)
plt.savefig('Linear_Reg_Engine_size')
```

### D. K-Means Clustering

```python
# import k-Means libraries
from sklearn.cluster import KMeans
```

```python
car_df.head()
```

```python
# Extract the input features
feature_names = ["Engine size", "Price"]
x = car_df[feature_names]
# Scale the input features
scale = StandardScaler()
scale.fit(x)
x_scaled = scale.transform(x)
```

```python
# Create a list to record the inertia
inertia = []
# Define the range of values of k
# that we will use.
k_range = range(1, 11)
# Loop through possible values of k
for k in k_range:
    # Initialise the model
    kmeans = KMeans(n_clusters = k, random_state = 45)
    # Fit the model to the scaled data
    kmeans.fit(x_scaled)
    # Record the inertia
    inertia.append(kmeans.inertia_)
```

```python
# Use matplotlib to plot the inertia versus k
plt.plot(k_range, inertia, 'bx-')
# Label the axes
plt.xlabel("Number of clusters")
plt.ylabel("Inertia")
# Show the figure
plt.show()
kmeans_final = KMeans(n_clusters = 3, random_state = 42)
```

```python
kmeans_final.fit(x_scaled)
# Use this fit to predict the cluster labels for each data point in the dataset.
cluster_labels_pred = kmeans_final.predict(x_scaled)
```

```python
# Import the internal measures
from sklearn.metrics import davies_bouldin_score, silhouette_score
# Calculate the Davies-Bouldin index
db_score = davies_bouldin_score(x_scaled, cluster_labels_pred)
# Calculate the Silhouette coefficient
s_score = silhouette_score(x_scaled, cluster_labels_pred)
# Print the results
print("Davies Bouldin Index: %.4f" % (db_score, ))
print("Silhouette Coefficient: %.4f" % (s_score, ))
```

```python
sns.scatterplot(data = car_df, x = "Engine size", y = "Price", hue =cluster_labels_pred)
# Add a legend to show the cluster labels
plt.legend(loc = "lower right")
# Label both axes
plt.xlabel("Engine size")
plt.ylabel("Car Price")
# Show the figure
plt.show()
```
